### cursor.fetchall(): 
This part of the code fetches all the rows from the result set obtainedfrom a database query using the cursor object. The fetchall() method retrieves all therows as a list of tuples.

### columns 
=[desc[0] for desc in cursor.description]: This part creates a list of columnnames for the DataFrame. It uses a list comprehension to iterate over thecursor.description, which is a list of 7-item tuples describing the columns in the resultset. The [desc[0] for desc in cursor.description] extracts the first element (columnname) from each tuple in the cursor.description and creates a list of column names.

### pd.DataFrame(...):
This part creates a Pandas DataFrame using the pd.DataFrame()constructor. It takes the fetched data (result of the query) and the list of column namesas arguments. The DataFrame is assigned to the variable customer_data.
In summary, this line of code fetches data from a database using a cursor, extractscolumn names from the cursor description, and creates a Pandas DataFrame with thefetched data and column names. The resulting DataFrame (customer_data) can be usedfor further analysis or manipulation using Pandas functionalities.

## Exploratory Data Analysis (EDA) and Visualization:

### ● Customer Analysis:

###  Identify the total number of customers City wise.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb=mydbconnect()
sqlstmt="select *  from customer" 
sql_query = pd.read_sql_query(sqlstmt,mydb)

df = pd.DataFrame(sql_query, columns = ['customer_id','name ','city','email','phone_no','address','pin_code'])
# Count the number of customers city-wise
city_wise_customer_count = df['city'].value_counts()

# Print the city-wise customer count
print(city_wise_customer_count)

# Plotting city-wise customer count
city_wise_customer_count.plot(kind='bar', color='purple', figsize=(8,5))
plt.title('Number of Customers City-wise')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.show()
```

### Identify the most frequent customers based on their order history.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb=mydbconnect()
sqlstmt="select *  from order_details" 
sql_query = pd.read_sql_query(sqlstmt,mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'customer_id', 'product_id', 'quantity', 'total_price', 'payment_mode', 'order_date', 'order_status'
])
# Count the number of orders for each customer
customer_order_counts = df['customer_id'].value_counts()

# Print the customer order counts
print(customer_order_counts)

# Plotting customer order counts
customer_order_counts.plot(kind='bar', color='teal', figsize=(8,5))
plt.title('Number of Orders for Each Customer')
plt.xlabel('Customer ID')
plt.ylabel('Number of Orders')
plt.show()
```

### Calculate the average, maximum, and minimum selling prices for products.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb=mydbconnect()
sqlstmt="select *  from product" 
sql_query = pd.read_sql_query(sqlstmt,mydb)

df = pd.DataFrame(sql_query, columns = ['product_id','product_name',' category ','sub_category',' original_price','selling_price','stock'])
average_price = df['selling_price'].mean()
maximum_price = df['selling_price'].max()
minimum_price = df['selling_price'].min()

# Print the results
print(f'Average Selling Price: {average_price}')
print(f'Maximum Selling Price: {maximum_price}')
print(f'Minimum Selling Price: {minimum_price}')

plt.figure(figsize=(8,5))
plt.bar(['Average', 'Maximum', 'Minimum'], [average_price, maximum_price, minimum_price], color=['red', 'teal', 'purple'])
plt.title('Average, Maximum, and Minimum Selling Prices')
plt.xlabel('Statistic')
plt.ylabel('Selling Price')
plt.show()
```

## ● Sales Analysis:

### Calculate total revenue generated from orders product wise

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.quantity, p.product_name, p. selling_price from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'product_name', 'quantity', 'selling_price'])
df['revenue'] = df['quantity'] * df['selling_price']
df.groupby('product_name')['revenue'].sum().sort_values(ascending=False).plot(kind='bar', xlabel='Product Name', ylabel='Total Revenue', title='Total Revenue Generated from Orders Product Wise', color='pink')
plt.show()
```

### Calculate total revenue product category wise percentage.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.quantity, p.product_name, p.selling_price, p.category from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'product_name', 'quantity', 'selling_price', 'category'])
df['revenue'] = df['quantity'] * df['selling_price']
revenue_by_category = df.groupby('category')['revenue'].sum()
total_revenue = revenue_by_category.sum()
revenue_percentage = (revenue_by_category / total_revenue) * 100

# Define a list of colors
colors = ['red', 'pink', 'purple']

# Pass the colors to the pie chart
revenue_percentage.plot(kind='pie', colors=colors, autopct='%1.1f%%')
plt.title('Total Revenue Generated from Orders Category Wise (in percentage)')
plt.ylabel('')
plt.show()

```

### Identify the most profitable products based on the difference between original and selling prices.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.quantity, p.product_name, p.selling_price, p.original_price from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'product_name', 'quantity', 'selling_price', 'original_price'])
df['profit'] = df['quantity'] * (df['selling_price'] - df['original_price'])
profit_by_product = df.groupby('product_name')['profit'].sum().sort_values(ascending=False)

profit_by_product.plot(kind='bar', xlabel='Product Name', ylabel='Profit', title='Profit Generated from Each Product', color='blue')
plt.show()
```

## ● Customer Order Patterns:

### Identify product names with the highest and lowest order quantities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.quantity, p.product_name from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'product_name', 'quantity'])
quantity_by_product = df.groupby('product_name')['quantity'].sum().sort_values(ascending=False)

# Product with highest order quantity
highest_quantity_product = quantity_by_product.idxmax()
print(f"The product with the highest order quantity is: {highest_quantity_product}")

# Product with lowest order quantity
lowest_quantity_product = quantity_by_product.idxmin()
print(f"The product with the lowest order quantity is: {lowest_quantity_product}")

# Plotting the products with highest and lowest order quantities
quantity_by_product[[highest_quantity_product, lowest_quantity_product]].plot(kind='bar', xlabel='Product Name', ylabel='Order Quantity', title='Order Quantity for Products with Highest and Lowest Quantities', color=['purple', 'teal'])
plt.show()
```

## ● Time-based Analysis:

### Month wise total sales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import calendar

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.order_date, o.quantity, p.product_name, p.selling_price from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'order_date', 'product_name', 'quantity', 'selling_price'])
df['order_date'] = pd.to_datetime(df['order_date'])  # Convert order_date to datetime
df['month'] = df['order_date'].dt.month  # Extract month from order_date
df['month'] = df['month'].apply(lambda x: calendar.month_name[x])  # Convert month number to month name

df['revenue'] = df['quantity'] * df['selling_price']
df_monthly_revenue = df.groupby('month')['revenue'].sum()  # Group by month and calculate total revenue

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # Define a list of colors
plt.bar(df_monthly_revenue.index, df_monthly_revenue, color=colors)
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Month-wise Total Revenue Generated from All Orders')
plt.show()
```

### Month and year wise total sales

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import calendar

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.order_date, o.quantity, p.product_name, p.selling_price from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'order_date', 'product_name', 'quantity', 'selling_price'])
df['order_date'] = pd.to_datetime(df['order_date'])  # Convert order_date to datetime
df['month_year'] = df['order_date'].dt.to_period('M')  # Extract month and year from order_date
df['month_year'] = df['month_year'].apply(lambda x: x.strftime('%B-%Y'))  # Convert month number to month name and concatenate with year

df['revenue'] = df['quantity'] * df['selling_price']
df_monthly_revenue = df.groupby('month_year')['revenue'].sum()  # Group by month and year and calculate total revenue

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # Define a list of colors
colors = colors * (len(df_monthly_revenue) // len(colors)) + colors[:len(df_monthly_revenue) % len(colors)]  # Repeat colors if necessary

plt.bar(df_monthly_revenue.index, df_monthly_revenue, color=colors)
plt.xlabel('Month and Year')
plt.ylabel('Total Revenue')
plt.title('Month and Year-wise Total Revenue Generated from All Orders')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```

### Identify peak order date

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import calendar

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.order_date, o.quantity, p.product_name, p.selling_price from product p natural join order_details o" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'order_date', 'product_name', 'quantity', 'selling_price'])
df['order_date'] = pd.to_datetime(df['order_date'])  # Convert order_date to datetime

df_order_count = df.groupby('order_date')['order_id'].count()  # Group by order_date and count the number of orders

peak_order_date = df_order_count.idxmax()  # Identify the date with the highest number of orders

print(f"The peak order date is {peak_order_date.strftime('%B %d, %Y')}, with a total of {df_order_count[peak_order_date]} orders")

# Plotting the trend graph 
plt.figure(figsize=(10, 5))
plt.plot(df_order_count.index, df_order_count, color='brown')
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.title('Trend of Number of Orders Over Time')
plt.grid(True)
plt.show() 
```

## Geographical Analysis:

### Explore the distribution of customers across different cities

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mydb = mydbconnect()
sqlstmt = "select distinct o.customer_id, c.city from order_details o join customer c on o.customer_id = c.customer_id" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['customer_id', 'city'])

df_customer_distribution = df['city'].value_counts()  # Count the number of customers from each city

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # Define a list of colors
colors = colors * (len(df_customer_distribution) // len(colors)) + colors[:len(df_customer_distribution) % len(colors)]  # Repeat colors if necessary

plt.figure(figsize=(10, 5))
plt.bar(df_customer_distribution.index, df_customer_distribution, color=colors)
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customers Across Different Cities')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```

### Analyze whether certain products or categories are more popular in a specific city.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mydb = mydbconnect()
sqlstmt = "select o.order_id, o.customer_id, p.product_name, p.category, c.city from product p natural join order_details o join customer c on o.customer_id = c.customer_id" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_id', 'customer_id', 'product_name', 'category', 'city'])

df_product_distribution = df.groupby(['city', 'category'])['order_id'].count().unstack()  # Count the number of orders for each category from each city

plt.figure(figsize=(10, 5))
sns.heatmap(df_product_distribution, cmap='viridis', annot=True, fmt='.2f')  # Use the 'viridis' colormap
plt.xlabel('Category')
plt.ylabel('City')
plt.title('Distribution of Product Categories Across Different Cities')
plt.show()
```

## ● Payment Analysis:

### Display successful and pending payments order counts.

```python
import pandas as pd
import matplotlib.pyplot as plt

mydb = mydbconnect()
sqlstmt = "SELECT o.order_status, COUNT(o.order_id) as number_of_orders FROM order_details o WHERE o.order_status IN ('Delivered', 'Pending','Not delivered') GROUP BY o.order_status" 
sql_query = pd.read_sql_query(sqlstmt, mydb)

df = pd.DataFrame(sql_query, columns = ['order_status', 'number_of_orders'])

# Define a color map
color_map = {'Delivered': 'green', 'Pending': 'orange', 'Not delivered': 'red'}

plt.figure(figsize=(10, 5))
# Use the color map for the bars
plt.bar(df['order_status'], df['number_of_orders'], color=[color_map[i] for i in df['order_status']])
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Order Status')
plt.show()
```
